## data clean

```python
plt.hist(df['Cycle time'], range=(58,64), bins=20, rwidth=.8)
plt.xlabel('cycle time')
plt.ylabel('frequency')
plt.title('fraucia product',fontsize=20)
```

```python
def count_rows(rows):
    return len(rows)
```

```python
by_cyc = df.groupby('Cycle time').apply(count_rows)
by_cyc
```

```python
df['Quality'].value_counts()
```

```python
by_quality = df.groupby('Quality').apply(count_rows)
```

```python
plt.style.use('ggplot') 
by_quality.plot(kind='barh', y="frequence", x="Quality", title="calculate of quality") 
```

```python
#set size of chart
fig = plt.figure(figsize=(10,10))
#"111" means "1x1 grid, first subplot" and "234" means "2x3 grid, 4th subplot".
ax = fig.add_subplot(111)
#parameter of bar
cax = ax.matshow(df.corr(), vmin=-1, vmax=1)
fig.colorbar(cax)
#intial the name of x and y, from 0 to 13, interval equal 1
ticks = np.arange(0,8,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
```
